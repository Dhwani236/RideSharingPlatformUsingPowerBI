```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px


np.random.seed(42)

n=500

cities=['Ahemdabad','Rajkot','Vadodara','Surat','Jamnagar']
area=['Urban','Semi-Urban']
usertype=['Regular','Frequent','Ocassional']
ridetype=['Mini','Sedan','SUV','Auto']
ridestatus=['Arrived','Cancelled']
paymentmode=['UPI','Cash','Card','Wallet']

data={
    "Ride_ID":np.arange(1001, 1001+n),
    "Customer_ID":range(1,1+n),
    "Driver_ID":range(5001, 5001+n),
    "City":np.random.choice(cities,n),
    "Area":np.random.choice(area,n),
    "Ride_Type":np.random.choice(ridetype,n),
    "Ride_Distance":np.random.randint(2,25,n),
    "Trip_Duration":np.random.randint(10,60,n),
    "Ride_Status":np.random.choice(ridestatus,n),
    "Payment_Mode":np.random.choice(paymentmode,n),
    "Ride_Rating":np.random.randint(1,5,n),
    "Ride_Wait_Time":np.random.randint(2,60,n)
}

df=pd.DataFrame(data)

df.to_csv("ride_sharing_platform.csv",sep=',',index=False,header=True)

# Base Fare=300
# Per Km Fare=100
   


```
